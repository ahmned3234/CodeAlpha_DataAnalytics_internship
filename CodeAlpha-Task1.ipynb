```python
# Imports
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
```

```python
url = "https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&q=Ai%20Engineer&start=0"

response = requests.get(url)

jobs_soup = BeautifulSoup(response.text,'html.parser') 

jobs_soup
```

```python
jobs_soup.title.text
```

```python
jobs = jobs_soup.find_all('div',{'class': 'css-laomuu'})
jobs[0]
```

```python
scrapped_jobs = []
for job in jobs :
    print(job.find('a' , {'class' : 'css-o171kl'}).text)
    print(job.find('a' , {'class' : 'css-o171kl'}).attrs['href'])
    scrapped_jobs.append({'Title': job.find('a' , {'class' : 'css-o171kl'}).text,'Link' : job.find('a' , {'class' : 'css-o171kl'}).attrs['href']
                         ,'Company': job.find('a', {'class' : 'css-17s97q8'}).text,'Location' : job.find('span', {'class' : 'css-5wys0k'}).text})
len(scrapped_jobs)
```

```python
df = pd.DataFrame(data = scrapped_jobs,columns = ['Title','Link','Company','Location'])
df.to_csv('jobs.csv')
df
```
